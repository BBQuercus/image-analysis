In order to test the different cell lines (Hela, HEK, U2OS each with either OptoG-mCherry or OptoG-mRuby) for their inducibiliy and overall handling. All cell lines were images under various conditons:

- No stress
- 1 hour of sodium arsenite stress (0.5 mM)
- 1 hour of induction with blue LED light
- 2 hours of induction with blue LED light
- 6 hours of induction with blue LED light
- 6 hours of induction with blue LED light with sodium arsenite added in the last hour

The follwing workflow takes an image from two channels (DAPI and mCherry or mRuby) and segments the DAPI channel into individual cells. These segments will then be analyzed in the second channel to measure the quantity of granules and total cellular area. The number of granules obtained will be used to determine the optimal induction condition and best cell line to continue with further experiments.

Lastly, some of the subsequent code for cellular segmentation was adapted from the European Molecular Biology Laboratory, Jonas Hartmann, Karin Sasaki, Toby Hodges (© 2018). The add_stat_annot function which proved to be extremely useful is borrowed from Marc Weber (© 2019) – thanks for these awesome 290 lines of code. Everything is distributed under a MIT license.

## Pipeline

### Import images

For this experiment, two wavelengths were recorded:

- Image 0 – DAPI 405 nm
- Image 1 – Granules 561 nm

The **file naming convention** to easily split into cell line, condition, transgene is: Cell-Transgene-Condition-Filenumber-Image Export-Number_z0?c?_ORG.tif

As all example files were renamed, the output of the batch processing is not shown.

```python
# File path-
root = 'optog_induction_example'
outfolder = 'optog_induction_example'

img_col_0 = np.zeros((1002, 1004))
img_col_1 = np.zeros((1002, 1004))

# Import all images in root
for file in os.listdir(root):
    if not file.endswith('.tif'):
        continue
    _img = imread(os.path.join(root, file))
    if 'c2' in file: # DAPI
        img_col_0 = np.dstack((img_col_0, _img))
    if 'c1' in file: # Granule
        img_col_1 = np.dstack((img_col_1, _img))

# Maximum projection and stacking of images
img0 = np.max(img_col_0, axis=2)
img1 = np.max(img_col_1, axis=2)
img = np.stack((img0, img1), axis=0)
img = img.astype(np.uint16)

''' Used for analysis
# Segment file name
filename = os.path.basename(os.path.normpath(root))
filename_cell = re.search(r'^(.*?)[CR]', filename).group(0)
filename_condition = re.findall(r'\-(..)\-', filename)[0]
filename_out = ''.join([filename, '.csv'])
'''

filename = 'example'
filename_cell = 'cell'
filename_condition = 'cond'
filename_out = ''.join([filename, '.csv'])


# Check if images are loaded correctly
fig, ax = plt.subplots(1, 2, figsize=(9, 9))
ax[0].imshow(img[0], interpolation=None, cmap='gray')
ax[0].set_title('DAPI channel')
ax[1].imshow(img[1], interpolation=None, cmap='gray')
ax[1].set_title('Granule channel')
plt.show()
```

<a id='section1.2'></a>
### Preprocessing 

To obtain a better segmentation later on, the image acquired through the DAPI channel will be smoothed by a gauss filer. Different sigmas (below) can be used to change 'smoothing intensity'.

```python
# Gauss-smoothen image
img0_sigma = 3
img0_smooth = ndi.filters.gaussian_filter(img[0], img0_sigma)

# Visualization
fig, ax = plt.subplots(1, 3, figsize=(10,10))
ax[0].imshow(img[0], interpolation='none', cmap='gray')
ax[0].set_title('Raw image')
ax[1].imshow(img0_smooth, interpolation='none', cmap='gray')
ax[1].set_title('Smoothed image')
ax[2].imshow(img0_smooth[400:600, 400:600], interpolation='none', cmap='gray')
ax[2].set_title('Zoomed')
plt.show()
```

<a id='section1.3'></a>
### Threshold

A global threshold will be applied to remove background noise and to be left with the DAPI nuclei. A variety of thresholding options can be selected. 

```python
# Decide on a thresholding option, here: otsu
img0_thresh = flt.threshold_otsu(img0_smooth)
img0_mem = img0_smooth > img0_thresh

# Visualization
plt.figure(figsize=(4,4))
plt.imshow(img0_mem, interpolation='none', cmap='gray')
plt.title('Thresholded DAPI')
plt.show()
```

<a id='section1.4'></a>
### Connected component labeling

As some DAPI nuclei currently show up to be combined, a further smoothing and a exact euclidean distance transform is used to separate these combined nuclei. Thereby, undersegmentation artifacts can be minimized.

```python
# Labeling of thresholded DAPI signal, smoothing
img0_cell_labels, _ = ndi.label(img0_mem)
img0_dist_trans = ndi.distance_transform_edt(img0_mem)
img0_dist_trans_smooth = ndi.filters.gaussian_filter(img0_dist_trans, sigma=3)

# Visualization
fig, ax = plt.subplots(1, 2, figsize=(10, 10))
ax[0].imshow(img0_cell_labels, interpolation='none', cmap='viridis')
ax[0].set_title('Labeled cells')
ax[1].imshow(img0_dist_trans_smooth, interpolation='none', cmap='viridis')
ax[1].set_title('Smoothed labels')
plt.show()
```

<a id='section1.5'></a>
### Labeling seeds

Local maxima are used to detect every DAPI nucleus. This provides a seed for the later segmentation.

```python
# Find local maxima – label seeds
img0_seeds = peak_local_max(img0_dist_trans_smooth, indices=False, min_distance=5)
img0_seeds_dil = ndi.filters.maximum_filter(img0_seeds, size=10) # Dilates seeds making them more visible
img0_seeds_labeled = ndi.label(img0_seeds)[0]
img0_seeds_labeled_dil = ndi.filters.maximum_filter(img0_seeds_labeled, size=10)

# Visualization
fig, ax = plt.subplots(1, 2, figsize=(10, 10))
ax[0].imshow(img0_smooth, interpolation='none', cmap='gray')
ax[0].imshow(np.ma.array(img0_seeds_dil, mask=img0_seeds_dil==0), interpolation='none', cmap='autumn')
ax[0].set_title('All seeds')
ax[1].imshow(img0_smooth, interpolation='none', cmap='gray')
ax[1].imshow(np.ma.array(img0_seeds_labeled_dil, mask=img0_seeds_labeled_dil==0), interpolation='none', cmap='prism')
ax[1].set_title('Labeled seeds')
plt.show()
```

<a id='section1.6'></a>
### Watershed / random walker and removal of border cells

Segmentation using the generated seeds. Inverted watershed or random walker can be used (change name). As cells on the image border aren't complete and might skew the later measurements, all border cells are removed. The currently used 'otsu' can be replaced with random_walker for equally nice results.

```python
# Use random_walker algorithm (inverted) to segment cells
img0_seg = watershed(~img0_smooth, img0_seeds_labeled)

# Create image border mask
img0_seg_clean = np.copy(img0_seg)
img0_border_mask = np.zeros(img0_seg.shape, dtype=np.bool)
img0_border_mask = ndi.binary_dilation(img0_border_mask, border_value=1)

# Delete border cells and re-label the remainder
for cell_ID in np.unique(img0_seg):
    img0_cell_mask = img0_seg==cell_ID 
    img0_cell_border_overlap = np.logical_and(img0_cell_mask, img0_border_mask)
    img0_total_overlap_pixels = np.sum(img0_cell_border_overlap)
    if img0_total_overlap_pixels > 0: 
        img0_seg_clean[img0_cell_mask] = 0
for new_ID, cell_ID in enumerate(np.unique(img0_seg_clean)[1:]):
    img0_seg_clean[img0_seg_clean==cell_ID] = new_ID + 1   
    
# Visualization
plt.figure(figsize=(4, 4))
plt.imshow(img0_smooth, interpolation='none', cmap='gray')
plt.imshow(np.ma.array(img0_seg_clean, mask=img0_seg_clean==0), interpolation='none', cmap='prism', alpha=0.4)
plt.title('Segmented cells')
plt.show()
```

<a id='section1.7'></a>
### Spot detection (global to find suitable parameters)

The most important step – find optimal parameters to detect spots in the second image channel. Use the widgets below to slide and find suitable values which can be further verified.

```python
# Find optimal parameters for detection
img1_sigma = 0
img1_block_size = 5
img1_thresh_area = 3000
img1_thresh_abs = 8000
img1_min_distance = 5

def test_spot_counter(image, block_size, threshold, min_distance):
    # Set thresholds to reduce noise
    thresh = flt.threshold_local(image, block_size, offset=0) 
    # Find and label seeds
    seeds = peak_local_max(thresh,
                           indices=False,
                           min_distance=min_distance,
                           threshold_abs=img1_thresh_abs)
    seeds_labeled = ndi.label(seeds)[0]
    seeds_labeled_dil = ndi.filters.maximum_filter(seeds_labeled, size=7) # Dilates seeds – more visible
    seeds_masked_labeled = np.ma.array(seeds_labeled_dil, mask=seeds_labeled_dil==0)
    seeds_unique = len(np.unique(seeds_labeled_dil)[1:])
    # Visualization
    print('Unique spots:', seeds_unique)
    plt.figure(figsize=(10, 10))
    plt.imshow(image, interpolation='none', cmap='gray')
    plt.imshow(seeds_masked_labeled, interpolation='none', cmap='prism', alpha=0.5)
    plt.title('Detected spots')
    
img1_smooth = ndi.filters.gaussian_filter(img[1], img1_sigma)
test_spot_counter(img1_smooth, img1_block_size, img1_thresh_abs, img1_min_distance)
```

<a id='section1.8'></a>
### Cellular spot detection

Inside generated cell segments, the cell area (using second channel) and the number of granules will be counted. As background intensity can vary, the optimal values can be evaluated using an @interact below or through trial and error in the second pannel.

```python
def spot_counter(image, block_size=5, threshold=8000, min_distance=0):
    # Set thresholds to reduce noise
    thresh_local = flt.threshold_local(image, block_size, offset=5) 
    thresh_global = thresh_local > threshold
    thresh_combined = np.ma.array(image, mask=thresh_global==0)
    # Find and label seeds
    seeds = peak_local_max(thresh_combined, indices=False, min_distance=min_distance)
    seeds_dil = ndi.filters.maximum_filter(seeds, size=2)
    seeds_masked = np.ma.array(seeds_dil, mask=seeds_dil==0)
    seeds_labeled = ndi.label(seeds)[0]
    seeds_labeled_dil = ndi.filters.maximum_filter(seeds_labeled, size=2)
    seeds_unique = len(np.unique(seeds_labeled_dil)[1:])
    return seeds_unique

# Parameters for detection
img1 = img[1]
img1_sigma_all = [0, 1] # Mean used for analysis
img1_block_size = 5
img1_thresh_abs  = [8000, 9000, 10000, 11000] # Mean used for analysis
img1_min_distance = 0
img1_thresh_area = 3000

results = {'filename'           : [],
           'sigma'              : [],
           'thresh_local'       : [],
           'block_size'         : [],
           'min_distance'       : [],
           'cell_line'          : [],
           #'transgene'          : [], for analysis
           'condition'          : [],
           'cell_ID'            : [],
           'cell_area'          : [],
           'cell_intensity_mean': [],
           'cell_intensity_std' : [],
           'spots'              : []}

for img1_sigma in img1_sigma_all:
    for img1_thresh in img1_thresh_abs:
        img1_smooth = ndi.filters.gaussian_filter(img[1], img1_sigma)

        for cell_ID in np.unique(img0_seg_clean)[1:]:
            img0_cell_mask = img0_seg_clean==cell_ID
            img1_cell = np.where(np.ma.array(img0_cell_mask, mask=img0_cell_mask==0), img1_smooth, 0)
            img1_cell_area = np.count_nonzero(img1_cell > img1_thresh_area)
            img1_cell_intensity_mean = np.mean(np.nonzero(img1_cell))
            img1_cell_intensity_std = np.std(np.nonzero(img1_cell))
            img1_cell_granules = spot_counter(img1_cell, img1_block_size, img1_thresh, img1_min_distance)

            results['filename'].append(os.path.basename(os.path.normpath(root)))
            results['cell_line'].append(filename_cell)
            #results['transgene'].append(filename_transgene) (for analysis)
            results['condition'].append(filename_condition)
            results['sigma'].append(img1_sigma)
            results['thresh_local'].append(img1_thresh)
            results['block_size'].append(img1_block_size)
            results['min_distance'].append(img1_min_distance)
            results['cell_ID'].append(cell_ID)
            results['cell_area'].append(img1_cell_area)
            results['cell_intensity_mean'].append(int(img1_cell_intensity_mean))
            results['cell_intensity_std'].append(int(img1_cell_intensity_std))
            results['spots'].append(img1_cell_granules)
```

```python
from scipy.ndimage.filters import maximum_filter
from scipy.ndimage.morphology import generate_binary_structure, binary_erosion

def detect_peaks(image):
    neighborhood = generate_binary_structure(2, 2)
    local_max = maximum_filter(image, footprint=neighborhood)==image
    background = (image==0)
    eroded_background = binary_erosion(background, structure=neighborhood, border_value=1)
    detected_peaks = local_max ^ eroded_background
    return detected_peaks

plt.figure(figsize=(10, 10))
thresh_local = flt.threshold_local(img1, 7, offset=0) 
thresh_global = thresh_local > 5000
thresh_combined = np.ma.array(img1, mask=thresh_global==0)
detected_peaks = detect_peaks(thresh_combined)
plt.imshow(img1, cmap='gray')
plt.imshow(detected_peaks, alpha=0.5)
plt.show()
```
