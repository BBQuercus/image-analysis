## Data analysis

### Reading output files

Reading all out-files.csv and combining them into one pandas dataframe. These are then cleaned up as follows:
* Select Minimum distance of 0
* Select Cell areas > 250 (to sort out small wrongly segmented cells)
* Select Mean of cell-intensity < 90% of highest mean (to sort out oversaturated cells)

```python
# Used for actual analyis
# Load all csv's into panda frame
outfolder = 'optog_induction_example'
files = glob.glob(os.path.join(outfolder, "*.csv"))

# Import all csv's
df_files = (pd.read_csv(f, sep="\t") for f in files)
df = pd.concat(df_files)
```

```python
# Used for example .csv file
f = 'merged.csv'
df = pd.read_csv(f, sep='\t', header=0)
df.head()
```

```python
# Data cleanup
df[['cell_area', 'spots', 'cell_intensity_mean', 'cell_intensity_std']] = df[['cell_area', 'spots', 'cell_intensity_mean', 'cell_intensity_std']].apply(pd.to_numeric)
df[['cell_line', 'transgene', 'condition']] = df[['cell_line', 'transgene', 'condition']].astype('category')

df_clean = df.loc[(df.min_distance == 0) &
                  (df.cell_area > 250) &
                  #(df.spots > 0) &
                  (df.cell_intensity_mean < df.cell_intensity_mean.max() * 0.9)]
```

### Barplot – induction efficiency

```python
# Make new dataframe with number of cells with spots
df_new = pd.DataFrame()

for cell_line in df_clean.cell_line.unique():
    for transgene in df_clean.transgene.unique():
        for condition in df_clean.condition.unique():
            df_curr = df_clean.loc[(df_clean.cell_line == cell_line) &
                                  (df_clean.transgene == transgene) &
                                  (df_clean.condition == condition)]
            cells_with = len(df_curr[df_curr.spots > 2])
            cells_total = len(df_curr)
            induction = round((cells_with / cells_total) * 100)
            df_new = df_new.append({'cell' : ('_'.join([cell_line, transgene, condition])),
                            'cell_line' : cell_line,
                            'transgene' : transgene,
                            'condition' : condition,
                            'induction' : induction},
                          ignore_index=True)
```

```python
# Over all samples / cell line comparison
sns.set(style='ticks', context='talk', font='sans-serif')

cell_lines = ['Hela', 'HEK', 'U2OS']
order_cell = ['Hela_C_6h', 'Hela_R_6h', 'HEK_C_6h', 'HEK_R_6h', 'U2OS_C_6h', 'U2OS_R_6h']
#label_cell = ['Hela mCherry', 'Hela mRuby', 'HEK mCherry', 'HEK mRuby', 'U2OS mCherry', 'U2OS mRuby']

ax = sns.barplot(x='cell_line', y='induction', hue='transgene', 
            data=df_new[df_new.condition == '6h'], hue_order=['C', 'R'], order=cell_lines)
ax.set_xticklabels(cell_lines)
ax.tick_params(axis='x', length=0)
ax.legend

plt.xlabel(None)
plt.ylim(0, 50)
plt.yticks(np.arange(0, 51, step=10))
plt.ylabel('Cells with spots / %')
plt.legend(frameon=False, labels=['mCherry', 'mRuby'])
sns.despine(trim=True, bottom=True, left=False)
plt.show()
```

```python
# Direct statistical test in graph
sns.set(style='ticks',
        context='talk',
        palette='OrRd',
        font='Helvetica')
order = ["No", "1h", "2h", "6h", "As", "6A"]
labels = ["None", "1h Light", "2h Light", "6h Light", "NaAsO$_2$", "6h Light +\n NaAsO$_2$"]

df_test = df_clean.loc[(df_clean.cell_line == 'U2OS') & (df_clean.transgene == 'R')]

n = []
for i in order:
    df_curr = df_test[df_test.condition == i]
    n.append(len(df_curr.index))
x = df_test.condition
y = df_test.spots

ax = sns.barplot(x=x, y=y, order=order, capsize=.2)
ax.set(xlabel=None, ylabel='Spots per cell')
ax.set_xticklabels(labels, rotation=90)
ax.tick_params(axis='x', length=0)
add_stat_annotation(ax, data=df_test, x=x, y=y,
                    boxPairList=[(order[1], order[2]), (order[1], order[3])],
                    test='t-test_ind', textFormat='star', loc='outside', useFixedOffset=True, verbose=1)
sns.despine(trim=True, bottom=True)
#plt.autoscale()

for index, patch in enumerate(ax.patches):
    ax.text(patch.get_x(), 0.2, str(n[index]))
    
#plt.savefig
```

```python
# Student's t-test
from scipy.stats import ttest_ind
from scipy.stats import f_oneway

data1 = df_clean[(df_clean.cell_line == 'U2OS') &
                 (df_clean.transgene == 'R') &
                 (df_clean.condition == 'No')].spots
data2 = df_clean[(df_clean.cell_line == 'U2OS') &
                 (df_clean.transgene == 'R') &
                 (df_clean.condition == '1h')].spots

alpha = 0.0005
stat, p = ttest_ind(data1, data2)
print('Students t-test:\n'
      'T-statistic = {}\n'
      'Two-tailed p-value = {}'.format(stat, p))
print('Same distributions – fail to reject H0\n') if p > alpha else print('Different distributions – reject H0\n')

stat, p = f_oneway(data1, data2)
print('Analysis of Variance (ANOVA) test:\n'
      'F-value = {}\n'
      'Associated p-value = {}'.format(stat, p))
print('Same distributions – fail to reject H0') if p > alpha else print('Different distributions – reject H0')
```
